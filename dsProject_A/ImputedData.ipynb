```python
import pandas as pd
from sklearn.impute import SimpleImputer
from seaborn import heatmap
```

```python
df = pd.read_csv("C:/Users/Avery/Desktop/STAT3400/ieee-fraud-detection/MasterTrain.csv")




si = SimpleImputer(strategy='most_frequent',)

si.fit(df)
df_imputed = pd.DataFrame(si.transform(df))

df_imputed = pd.DataFrame(si.fit_transform(df),
             columns = si.get_feature_names_out())

```

```python
df_imputed.to_csv("MasterTrainSetImputed.csv",index=False)
```

```python
heatmap(df_imputed.isna())
```
